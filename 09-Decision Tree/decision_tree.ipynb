```python
X = pd.get_dummies(df.drop('species', axis=1), drop_first=True)
y = df.species
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.tree import DecisionTreeClassifier
dtree = DecisionTreeClassifier()
```

```python
dtree.fit(X_train, y_train)
base_preds = dtree.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
accuracy_score(y_test, base_preds)
```

```python
confusion_matrix(y_test, base_preds)
```

```python
print(classification_report(y_test, base_preds))
```

```python
dtree.feature_importances_
```

```python
pd.DataFrame(dtree.feature_importances_, index=X.columns, columns=['importance']).sort_values('importance', ascending=False)
```

```python
from sklearn.tree import plot_tree
plt.figure(figsize=(12,8), dpi=200)
plot_tree(dtree, filled=True, feature_names=X.columns);
```

```python
def report_model(model):
    preds = model.predict(X_test)
    print(classification_report(y_test, preds))
    print('Accuracy Score: ', accuracy_score(y_test, preds))
    print('\n')
    plt.figure(figsize=(12,8), dpi=200)
    plot_tree(model, filled=True, feature_names=X.columns);
```

```python
report_model(dtree)
```

```python
pruned_tree = DecisionTreeClassifier(max_depth=2)
pruned_tree.fit(X_train, y_train)
```

```python
report_model(pruned_tree)
```

```python
max_leaf_tree = DecisionTreeClassifier(max_leaf_nodes=3)
max_leaf_tree.fit(X_train, y_train)
```

```python
report_model(max_leaf_tree)
```

```python
entropy_tree = DecisionTreeClassifier(criterion='entropy')
entropy_tree.fit(X_train, y_train)
report_model(entropy_tree)
```
